```python
import pandas as pd

# Load market data
market_data = pd.read_csv(
    "../data/market_data_full.csv",
    parse_dates=["date"],
    index_col="date"
)

# Load macro events
macro_events = pd.read_csv(
    "../data/macro_events.csv",
    parse_dates=["event_date"]
)

print("Market data shape:", market_data.shape)
print("Macro events shape:", macro_events.shape)

market_data.head(), macro_events.head()
```

```python
event_windows = {
    "pre_1d": -1,
    "pre_5d": -5,
    "post_1d": 1,
    "post_5d": 5
}

event_windows
```

```python
def get_nearest_trading_day(date, market_index):
    if date in market_index:
        return date
    return market_index[market_index.get_indexer([date], method="nearest")[0]]
```

```python
event_reactions = []

for _, event in macro_events.iterrows():
    event_date = event["event_date"]
    event_type = event["event_type"]
    event_id = event["event_id"]

    aligned_date = get_nearest_trading_day(event_date, market_data.index)

    for label, offset in event_windows.items():
        window_date = aligned_date + pd.Timedelta(days=offset)

        if window_date in market_data.index:
            row = {
                "event_id": event_id,
                "event_type": event_type,
                "event_date": event_date,
                "window": label,
                "window_date": window_date,
            }

            for col in market_data.columns:
                row[col] = market_data.loc[window_date, col]

            event_reactions.append(row)

event_reactions_df = pd.DataFrame(event_reactions)
event_reactions_df.head()
```

```python
# Pivot so each event has windows side-by-side
pivot_cols = ["SP500", "NASDAQ", "DXY", "VIX", "GOLD", "UST_2Y", "UST_10Y"]

event_pivot = (
    event_reactions_df
    .pivot_table(
        index=["event_id", "event_type", "event_date"],
        columns="window",
        values=pivot_cols
    )
)

event_pivot.head()
```

```python
reactions = event_pivot.copy()

# % returns for risk assets
for asset in ["SP500", "NASDAQ", "DXY", "GOLD"]:
    reactions[(asset, "ret_1d")] = (
        event_pivot[(asset, "post_1d")] / event_pivot[(asset, "pre_1d")] - 1
    ) * 100

# Vol change (levels)
reactions[("VIX", "chg_1d")] = (
    event_pivot[("VIX", "post_1d")] - event_pivot[("VIX", "pre_1d")]
)

# Yield changes (bps)
for rate in ["UST_2Y", "UST_10Y"]:
    reactions[(rate, "bps_1d")] = (
        event_pivot[(rate, "post_1d")] - event_pivot[(rate, "pre_1d")]
    ) * 100

reactions.head()
```

```python
# Keep only the calculated reaction columns
reaction_cols = [col for col in reactions.columns if col[1] in ["ret_1d", "chg_1d", "bps_1d"]]

reaction_table = reactions[reaction_cols].reset_index()

reaction_table.head()
```

```python
import os

PROJECT_ROOT = os.path.dirname(os.getcwd())
OUT_PATH = os.path.join(PROJECT_ROOT, "outputs", "event_reaction_1d.csv")

reaction_table.to_csv(OUT_PATH, index=False)

print("Saved event reactions to:", OUT_PATH)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
avg_reactions = (
    reaction_table
    .groupby("event_type")
    .mean(numeric_only=True)
    .reset_index()
)

avg_reactions
```

```python
plt.figure(figsize=(10, 6))

for asset in ["SP500_ret_1d", "NASDAQ_ret_1d"]:
    plt.bar(
        avg_reactions["event_type"],
        avg_reactions[asset],
        label=asset.replace("_ret_1d", "")
    )

plt.title("Average 1-Day Equity Returns by Macro Event")
plt.ylabel("Return (%)")
plt.legend()
plt.show()
```
